```python
def sel():
    p1 = 0
    while p1 == 0:
        p1 = input('Player1 choose your symbol (X/O) ')
        if p1 == 'X':
            p2 = 'O'
        elif p1 == 'O':
            p2 = 'X'
        else:
            p1 = 0
            print('Please input correct symbol!')
    return p1, p2
        
```

```python
def init():
    ch = 0
    while ch == 0:
        ch = input('Are you both ready? (Y/N) ')
        if ch == 'Y':
            return True
        elif ch == 'N':
            return False
        else:
            ch = 0
            print('Sorry could not understand the input! Please try again.. ')
            
```
